## Simply Import and initialize Unifier and Go!

```python
from unifier import unifier as Unifier
```

```python
Unifier.user = 'unifier username'
Unifier.token = 'unifier api token'
os.environ['UNIFIER_USER'] = Unifier.user
os.environ['UNIFIER_TOKEN'] = Unifier.token
```

## Retrieve data for a specific date using the asof_date parameter

```python
df = Unifier.get_dataframe(name="xtech_us_eqt_flow_r3k_with_isin_1min_trial", asof_date='2024-01-18', limit=10)
display(df)
```

## Retrieve data for a specific date range using the back_to and up_to parameters


```python
df = Unifier.get_dataframe(name='xtech_us_eqt_flow_r3k_with_isin_1min_trial', back_to='2024-01-31', up_to='2024-02-01', limit=100)
display(df)
```

## Retrieve data for a specific date with a specific ticker (key='ticker') using the asof_date parameter

```python
df = Unifier.get_dataframe(name='xtech_us_eqt_flow_r3k_with_isin_1min_trial',key='AAPL', asof_date='2024-01-01', limit=100)
display(df)
```

## Retrieve data for a specific date range with a specific ticker (key='ticker') using the back_to and up_to parameters


```python
df = Unifier.get_dataframe(name='xtech_us_eqt_flow_r3k_with_isin_1min_trial',key='AAPL', back_to='2019-01-01', up_to='2024-06-01', limit=100)
display(df)
```

## Download data file for a specific date using asof_date


```python
Unifier.replicate(name="xtech_us_eqt_flow_r3k_with_isin_1min_trial", target_location="./data", asof_date='2026-01-13')
```
