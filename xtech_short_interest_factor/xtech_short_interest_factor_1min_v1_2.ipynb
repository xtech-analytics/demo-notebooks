```python
df =  Unifier.get_dataframe(name="xtech_us_eqt_short_interest_factor_1min_v1_2", back_to='2025-11-18', limit=10)
display(df)
```

```python
df = Unifier.get_dataframe(name='xtech_us_eqt_short_interest_factor_1min_v1_2', back_to='2021-01-11', up_to='2021-02-01', limit=10)
display(df)
```
