## Simply Import and initialize Unifier and Go!

```python
from unifier import unifier as Unifier
```

```python
Unifier.user = 'unifier username'
Unifier.token = 'unifier api token'
os.environ['UNIFIER_USER'] = Unifier.user
os.environ['UNIFIER_TOKEN'] = Unifier.token
```

## Retrieve data for a specific date range using the back_to and up_to parameters


```python
df = Unifier.get_dataframe(name='xtech_us_edi_eod_prices_adjusted_allcomposite_v2', back_to='2022-01-31', up_to='2024-02-01', limit=100)
display(df)
```

## Retrieve data for a specific date with a specific ticker (key='SecID') using the asof_date parameter

```python
df = Unifier.get_dataframe(name='xtech_us_edi_eod_prices_adjusted_allcomposite_v2',key='4909982', asof_date='2023-09-12', limit=100)
display(df)
```

## Retrieve data for a specific date range with a specific SecID (key='SecID') using the back_to and up_to parameters


```python
df = Unifier.get_dataframe(name='xtech_us_edi_eod_prices_adjusted_allcomposite_v2',key='4909982', back_to='2022-01-01', up_to='2024-06-01', limit=100)
display(df)
```

## Download data file for a specific date using asof_date


```python
Unifier.replicate(name="xtech_us_edi_eod_prices_adjusted_allcomposite_v2", target_location="./data", asof_date='2026-01-13')
```
