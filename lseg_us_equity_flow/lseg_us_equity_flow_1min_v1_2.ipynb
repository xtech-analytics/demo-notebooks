## Retrieve data for a specific date using the asof_date parameter

```python
df = Unifier.get_dataframe(name="lseg_us_equity_flow_1min_v1_2", asof_date='2024-01-18', limit=10)
display(df)
```

## Retrieve data for a specific date range using the back_to and up_to parameters


```python
df = Unifier.get_dataframe(name='lseg_us_equity_flow_1min_v1_2', back_to='2024-01-31', up_to='2024-02-01', limit=100)
display(df)
```
